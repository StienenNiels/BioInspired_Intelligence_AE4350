```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df1 = pd.read_csv('recordings/TimeSteps1e3_0.csv')
df2 = pd.read_csv('recordings/TimeSteps1e4_0.csv')
df3 = pd.read_csv('recordings/TimeSteps1e5_0.csv')
df4 = pd.read_csv('recordings/TimeSteps1e6_0.csv')

# Create a figure and a set of subplots with 6 subplots arranged vertically
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 10))

axes[0][0].plot(df1['cart_pos'])
axes[0][0].plot(df2['cart_pos'])
axes[0][0].plot(df3['cart_pos'])
axes[0][0].plot(df4['cart_pos'])
axes[0][0].set_title(f'Plot of cart position at each timestep')
axes[0][0].set_xlabel('Timestep')
axes[0][0].set_ylabel('Cart position')
axes[0][0].legend(['1e3 model', '1e4 model', '1e5 model', '1e6 model'])
axes[0][0].set_ylim((-1, 1))

axes[1][0].plot(df1['pole_angle'])
axes[1][0].plot(df2['pole_angle'])
axes[1][0].plot(df3['pole_angle'])
axes[1][0].plot(df4['pole_angle'])
axes[1][0].set_title(f'Plot of pole angle at each timestep')
axes[1][0].set_xlabel('Timestep')
axes[1][0].set_ylabel('Pole angle')
axes[1][0].legend(['1e3 model', '1e4 model', '1e5 model', '1e6 model'])
axes[1][0].set_ylim((-0.1, 0.1))

axes[2][0].plot(df1['reward'])
axes[2][0].plot(df2['reward'])
axes[2][0].plot(df3['reward'])
axes[2][0].plot(df4['reward'])
axes[2][0].set_title(f'Plot of reward at each timestep')
axes[2][0].set_xlabel('Timestep')
axes[2][0].set_ylabel('Reward')
axes[2][0].legend(['1e3 model', '1e4 model', '1e5 model', '1e6 model'])
# axes[2][0].set_ylim((-0.1, 0.1))

axes[0][1].plot(df1['cart_velocity'])
axes[0][1].plot(df2['cart_velocity'])
axes[0][1].plot(df3['cart_velocity'])
axes[0][1].plot(df4['cart_velocity'])
axes[0][1].set_title(f'Plot of cart velocity at each timestep')
axes[0][1].set_xlabel('Timestep')
axes[0][1].set_ylabel('Cart velocity')
axes[0][1].legend(['1e3 model', '1e4 model', '1e5 model', '1e6 model'])
axes[0][1].set_ylim((-1, 1))

axes[1][1].plot(df1['pole_ang_vel'])
axes[1][1].plot(df2['pole_ang_vel'])
axes[1][1].plot(df3['pole_ang_vel'])
axes[1][1].plot(df4['pole_ang_vel'])
axes[1][1].set_title(f'Plot of pole velocity at each timestep')
axes[1][1].set_xlabel('Timestep')
axes[1][1].set_ylabel('Pole velocity')
axes[1][1].legend(['1e3 model', '1e4 model', '1e5 model', '1e6 model'])
axes[1][1].set_ylim((-0.2, 0.2))

axes[2][1].plot(df1['total_reward'])
axes[2][1].plot(df2['total_reward'])
axes[2][1].plot(df3['total_reward'])
axes[2][1].plot(df4['total_reward'])
axes[2][1].set_title(f'Plot of cumulative reward')
axes[2][1].set_xlabel('Timestep')
axes[2][1].set_ylabel('Reward')
axes[2][1].legend(['1e3 model', '1e4 model', '1e5 model', '1e6 model'])
axes[2][1].set_ylim((0, 500))

x_limits = (0, 400)  # Define x-axis limits as a tuple (xmin, xmax)
for ax in axes.flat:
    ax.set_xlim(x_limits)

# Adjust the layout so the plots do not overlap
plt.tight_layout()
plt.savefig('report_images/'+'compare_timesteps.eps')

# Show the plots
plt.show()
```
