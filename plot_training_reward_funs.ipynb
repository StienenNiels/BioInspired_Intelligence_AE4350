```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df1 = pd.read_csv('logs/Default_Reward.csv')
df2 = pd.read_csv('logs/Pos_Reward.csv')
df3 = pd.read_csv('logs/Vel_Reward.csv')

# Create a new figure
plt.figure(figsize=(10, 6))  # Set the figure size

plt.plot(df1['episode']*48, df1['average_reward'], label='Default Reward')
plt.plot(df2['episode']*45, df2['average_reward'], label='Position Reward')
plt.plot(df3['episode']*53, df3['average_reward'], label='Velocity Reward')
plt.title(f'Plot of average reward per episode')
plt.xlabel('Episode')
plt.ylabel('Average Reward')
plt.legend()
plt.xlim(0, 1e5)

# Adjust the layout so the plots do not overlap
plt.tight_layout()
plt.savefig('report_images/'+'avg_reward.eps')

# Show the plots
plt.show()
```
