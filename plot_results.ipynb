```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
# Make sure to update 'your_file.csv' with the actual path to your CSV file
df = pd.read_csv('recordings/Default_0.csv')
df2 = pd.read_csv('recordings/Default_1.csv')
df3 = pd.read_csv('logs/Default.csv')

# Display the first few rows of the DataFrame to understand its structure
df.head()

# Create a figure and a set of subplots with 6 subplots arranged vertically
fig, axes = plt.subplots(nrows=8, ncols=1, figsize=(10, 20))

# Iterate through each column and plot it on a separate subplot
for i, column in enumerate(df.columns):
    if i != 0:
        j = i-1
        axes[j].plot(df[column])
        axes[j].plot(df2[column])
        axes[j].set_title(f'Plot of {column}')
        axes[j].set_xlabel('Index')
        axes[j].set_ylabel(column)

axes[7].plot(df3['episode'], df3['average_reward'], label='Reward per Episode')
axes[7].set_title(f'Plot of {df3.columns[1]}')
axes[7].set_xlabel('Episode')
axes[7].set_ylabel(df3.columns[1])

# Adjust the layout so the plots do not overlap
plt.tight_layout()

# Show the plots
plt.show()
```
