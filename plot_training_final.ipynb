```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df1 = pd.read_csv('logs/Final.csv')

# Create a new figure
plt.figure(figsize=(10, 6))  # Set the figure size

plt.plot(df1['episode']*500, df1['average_reward'], label='1e3 steps')
plt.title(f'Plot of average reward per episode')
plt.xlabel('Episode')
plt.ylabel('Average Reward')
plt.xlim(0, 1e6)

# Adjust the layout so the plots do not overlap
plt.tight_layout()
plt.savefig('report_images/'+'training_final.eps')

# Show the plots
plt.show()
```
